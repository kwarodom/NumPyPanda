#Challenge Objectives
1. given a list of words from a dictionary file, separate words into classes of words with the same length\
2. for each class of words of the same length, find all anagrams
3. count the total number of anagrams in each class

```python
import collections
```

```python
wordclean = sorted(list(set([word.strip().lower() for word in open('words','r')])))
```

```python
def signature(word):
    return ''.join(sorted(word))
```

```python
words_bysig = collections.defaultdict(list)

for word in wordclean:
    words_bysig[signature(word)].append(word)
```

```python
def anagram_fast(myword):
    return words_bysig[signature(myword)]
```

```python
anagrams_all = {word: anagram_fast(word) for word in wordclean if len(anagram_fast(word)) > 1}
```

```python
words_by_length = collections.defaultdict(list)
for word in wordclean:
    words_by_length[len(word)].append(word)
```

```python
anagrams_bylength = {}

for length, words in words_by_length.items():
    anagrams_bylength[length] = {word:anagram_fast(word) for word in words if len(anagram_fast(word))>1}
```
