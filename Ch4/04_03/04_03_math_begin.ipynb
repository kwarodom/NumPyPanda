```python
import numpy as np
import matplotlib.pyplot as pp
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,10,40)
```

```python
x
```

```python
sinx = np.sin(x)
```

```python
pp.plot(x,sinx)
```

```python
cosx = np.cos(x)
```

```python
pp.plot(x,sinx,'x')
pp.plot(x,cosx,'o')

pp.legend(['sin(x)','cos(x)'])
```

```python
y = sinx * cosx
z = cosx**2 - sinx**2

pp.plot(x,y)
pp.plot(x,z)

pp.legend(['sin(x) cos(x)','cos(x)^2 - sin(x)^2'])
```

```python
np.dot(sinx,cosx) # inner product of two vectors
```

```python
np.outer(sinx,cosx) # every possible combinations of the elements of two vectors
```

```python
v = np.linspace(0,10,5)
v + 1 #broadcasting
```

```python
vv = np.outer(v,v)
vv + v
```
